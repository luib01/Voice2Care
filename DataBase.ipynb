```python
import os
import json

from datetime import datetime
from pymongo import MongoClient
from pymongo.errors import PyMongoError

# Configurazione MongoDB
MONGO_URI = "mongodb://localhost:27017"
DB_NAME = "cartella_clinica_db"
COLLECTION_NAME = "interventi"

# Cartella contenente i file JSON
JSON_FOLDER = "./cartella_clinica"

# Connessione a MongoDB
client = MongoClient(MONGO_URI)
db = client[DB_NAME]
collection = db[COLLECTION_NAME]

# Conta i file caricati
inseriti = 0
skippati = 0

# Inserimento dei file JSON
for filename in os.listdir(JSON_FOLDER):
    if filename.endswith(".json"):
        filepath = os.path.join(JSON_FOLDER, filename)
        try:
            with open(filepath, "r", encoding="utf-8") as f:
                data = json.load(f)
                if isinstance(data, dict):  # Assicura che sia un oggetto JSON
                    collection.insert_one(data)
                    inseriti += 1
                else:
                    print(f"[⚠️] File non valido (non è un oggetto JSON): {filename}")
                    skippati += 1
        except (json.JSONDecodeError, PyMongoError) as e:
            print(f" Errore nel file {filename}: {e}")
            skippati += 1

print(f"\n Inserimento completato: {inseriti} file caricati, {skippati} file scartati.")

```
